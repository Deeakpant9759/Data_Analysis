```python
import os
import shutil
from pathlib import Path

# ðŸ“‚ Source folder
source_dir = Path(r"c:\Users\MICILMEDS\Documents\Medi_final\correct_data\Category\Drive_images")

# ðŸ“‚ Destination root (splitted batches will be created here)
dest_root = source_dir.parent / "Drive_images_split"
dest_root.mkdir(exist_ok=True)

# Batch size
batch_size = 1999

# Collect all image files (jpg, png, jpeg, etc.)
image_ext = {".jpg", ".jpeg", ".png", ".bmp", ".tiff", ".webp"}
all_images = [f for f in sorted(source_dir.iterdir())
              if f.suffix.lower() in image_ext and f.is_file()]

print(f"Total images found: {len(all_images)}")

for i in range(0, len(all_images), batch_size):
    batch_num = i // batch_size + 1
    batch_folder = dest_root / f"Batch_{batch_num:03d}"
    batch_folder.mkdir(exist_ok=True)
    print(f"Creating {batch_folder}")

    # Copy images to this batch folder without changing names
    for img_path in all_images[i:i + batch_size]:
        new_path = batch_folder / img_path.name  # Keep original name
        shutil.copy2(img_path, new_path)        # Use shutil.move(...) to move instead of copy
```
